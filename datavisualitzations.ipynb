```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import random
x=np.arange(-20,21)
```

```python
plt.figure(figsize=(5,5))#grafiğin en boy oranı
plt.title("Deneme Grafik")
plt.plot(x,x**2)
plt.plot(x,-1*x**2)
```

```python
plt.subplot(1, 2, 1)  # rows, columns, panel selected
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])

plt.xlabel('X')
plt.ylabel('X Squared')
```

```python
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My Nice Plot")

axes.legend()

fig
```

```python

fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid', label="solid")
axes.plot(x, x + 1, linestyle='dashed', label="dashed")
axes.plot(x, x + 2, linestyle='dashdot', label="dashdot")
axes.plot(x, x + 3, linestyle='dotted', label="dotted")

axes.set_title("My Nice Plot")
axes.legend()
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, '-og', label="solid green")
axes.plot(x, x + 1, '--c', label="dashed cyan")
axes.plot(x, x + 2, '-.b', label="dashdot blue")
axes.plot(x, x + 3, ':r', label="dotted red")

axes.set_title("My Nice Plot")

axes.legend()
```
